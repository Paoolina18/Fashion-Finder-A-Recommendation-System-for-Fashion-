```python
import pandas as pd
import os
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Paso 1: Crear la base de datos inicial con distribución específica
def create_dataframe():
    """
    Create a DataFrame with specific counts for tops, bottoms, and dresses.
    """
    # Ruta de las imágenes
    image_path = r'C:\Users\POspina\OneDrive - SLB\Desktop\Backup\Tesis\Fotos'

    # Generar datos para cada tipo de prenda
    data = []

    # Tops (51 items)
    for i in range(1, 52):
        data.append({
            'ID': i,
            'Type': 'Top',
            'Color': random.choice(['White', 'Black', 'Red', 'Blue']),
            'Size': random.choice(['S', 'M', 'L', 'XL']),
            'Image': os.path.join(image_path, f"Top ({i}).png")
        })

    # Bottoms (22 items)
    for i in range(1, 23):
        data.append({
            'ID': 51 + i,
            'Type': 'Bottom',
            'Color': random.choice(['White', 'Black', 'Red', 'Blue']),
            'Size': random.choice(['S', 'M', 'L', 'XL']),
            'Image': os.path.join(image_path, f"Bottom ({i}).png")
        })

    # Dresses (15 items)
    for i in range(1, 16):
        data.append({
            'ID': 73 + i,
            'Type': 'Dress',
            'Color': random.choice(['White', 'Black', 'Red', 'Blue']),
            'Size': random.choice(['S', 'M', 'L', 'XL']),
            'Image': os.path.join(image_path, f"Dress ({i}).png")
        })

    return pd.DataFrame(data)

# Crear el DataFrame inicial
df = create_dataframe()

# Paso 2: Depuración de rutas de imágenes
def debug_image_paths(df):
    """
    Check if all image paths in the DataFrame exist.
    """
    missing_images = [img for img in df['Image'] if not os.path.exists(img)]
    if missing_images:
        print("\nWarning: The following image paths are missing:")
        for img in missing_images:
            print(img)
    else:
        print("\nAll images are correctly found.")

debug_image_paths(df)

# Paso 3: Mostrar outfits en columnas
def display_outfits_in_columns(outfits):
    """
    Display three outfits in three columns: tops/dresses on the first row, bottoms/jackets on the second row.
    """
    fig, axes = plt.subplots(2, 3, figsize=(12, 8))  # 2 rows, 3 columns
    fig.suptitle("Outfits Proposal", fontsize=16, y=0.95)

    for i, (top_or_dress, bottom_or_jacket) in enumerate(outfits):
        # Load and display top or dress image
        if os.path.exists(top_or_dress['Image']):
            img = mpimg.imread(top_or_dress['Image'])
            axes[0, i].imshow(img)
            axes[0, i].set_title(f"{top_or_dress['Type']} ({top_or_dress['Color']}, {top_or_dress['Size']})")
        else:
            axes[0, i].text(0.5, 0.5, "Image not found", ha='center', va='center', fontsize=10, color='red')
            axes[0, i].set_title(f"{top_or_dress['Type']} ({top_or_dress['Color']}, {top_or_dress['Size']})")
        axes[0, i].axis('off')

        # Load and display bottom or jacket image
        if os.path.exists(bottom_or_jacket['Image']):
            img = mpimg.imread(bottom_or_jacket['Image'])
            axes[1, i].imshow(img)
            axes[1, i].set_title(f"{bottom_or_jacket['Type']} ({bottom_or_jacket['Color']}, {bottom_or_jacket['Size']})")
        else:
            axes[1, i].text(0.5, 0.5, "Jacket not available" if bottom_or_jacket['Type'] == "Jacket" else "Image not found", 
                            ha='center', va='center', fontsize=10, color='red')
            axes[1, i].set_title(f"{bottom_or_jacket['Type']} ({bottom_or_jacket['Color']}, {bottom_or_jacket['Size']})")
        axes[1, i].axis('off')

    plt.tight_layout(rect=[0, 0, 1, 0.9])  # Adjust layout to fit title
    plt.show()

# Paso 4: Generar y mostrar outfits
def display_outfits(df):
    """
    Display three randomly generated outfits and return the list of outfits.
    """
    outfits = []
    
    for _ in range(3):
        # Randomly select a top or dress
        top_or_dress = df[df['Type'].isin(['Top', 'Dress'])].sample(1).to_dict('records')[0]
        
        # If the selection is a dress, recommend a jacket
        if top_or_dress['Type'] == 'Dress':
            bottom_or_jacket = {
                'ID': None,
                'Type': 'Jacket',
                'Color': 'N/A',
                'Size': 'N/A',
                'Image': 'N/A'  # No images available for jackets yet
            }
        else:  # Otherwise, recommend a bottom
            bottom_or_jacket = df[df['Type'] == 'Bottom'].sample(1).to_dict('records')[0]

        outfits.append((top_or_dress, bottom_or_jacket))

    # Display outfits in a grid layout
    display_outfits_in_columns(outfits)

    return outfits

# Paso 5: Recolectar feedback
def collect_outfit_feedback(df, feedback_db, iterations):
    """
    Iterate through multiple feedback rounds and collect user input.
    Allows the user to regenerate outfits before making a final choice.
    """
    print(f"\nDisplaying {iterations} outfit sets to the user...\n")
    for i in range(iterations):
        print(f"Iteration {i + 1}:")
        
        while True:  # Loop to allow outfit regeneration
            outfits = display_outfits(df)  # Display the outfits
            
            # Ask if the user wants to regenerate outfits
            regenerate = input("Do you want to regenerate the outfits? (yes/no): ").strip().lower()
            if regenerate in ['no', 'n']:
                break  # Exit the loop if the user is satisfied
            elif regenerate not in ['yes', 'y']:
                print("Invalid input. Please type 'yes' or 'no'.")
        
        while True:  # Loop to get a valid outfit choice
            try:
                choice = int(input("Which of the 3 outfits did you like the most? (1, 2, or 3): ").strip())
                if choice in [1, 2, 3]:
                    break
                else:
                    print("Please choose a valid option (1, 2, or 3).")
            except ValueError:
                print("Invalid input. Please try again.")
        
        log_outfit_feedback(outfits, choice, feedback_db)

# Paso 6: Registrar feedback
def log_outfit_feedback(outfits, choice, feedback_db):
    """
    Log user's feedback for the selected outfit into the feedback database.
    """
    top_or_dress, bottom_or_jacket = outfits[choice - 1]  # Extract the selected outfit
    feedback_db.append({
        'Top_or_Dress_ID': top_or_dress['ID'], 
        'Bottom_or_Jacket_ID': bottom_or_jacket.get('ID'),  # Handle Jacket case
        'Feedback': True  # Mark feedback as positive
    })

# Ejecutar el sistema
feedback_db = []
collect_outfit_feedback(df, feedback_db, iterations=3)
print("\nFeedback recorded:")
print(pd.DataFrame(feedback_db))
```

```python
import pandas as pd
import os
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from ipywidgets import Button, VBox, HBox, Output, Layout

# Crear el DataFrame inicial con distribución específica
def create_dataframe():
    """
    Create a DataFrame with specific counts for tops, bottoms, and dresses.
    """
    image_path = r'C:\Users\POspina\OneDrive - SLB\Desktop\Backup\Tesis\Fotos'

    # Generate data for each type
    data = []

    # Tops (51 items)
    for i in range(1, 52):
        data.append({
            'ID': i,
            'Type': 'Top',
            'Color': random.choice(['White', 'Black', 'Red', 'Blue']),
            'Size': random.choice(['S', 'M', 'L', 'XL']),
            'Image': os.path.join(image_path, f"Top ({i}).png")
        })

    # Bottoms (22 items)
    for i in range(1, 23):
        data.append({
            'ID': 51 + i,
            'Type': 'Bottom',
            'Color': random.choice(['White', 'Black', 'Red', 'Blue']),
            'Size': random.choice(['S', 'M', 'L', 'XL']),
            'Image': os.path.join(image_path, f"Bottom ({i}).png")
        })

    # Dresses (15 items)
    for i in range(1, 16):
        data.append({
            'ID': 73 + i,
            'Type': 'Dress',
            'Color': random.choice(['White', 'Black', 'Red', 'Blue']),
            'Size': random.choice(['S', 'M', 'L', 'XL']),
            'Image': os.path.join(image_path, f"Dress ({i}).png")
        })

    return pd.DataFrame(data)

df = create_dataframe()
feedback_db = []

# Función para generar outfits
def generate_outfits(df):
    """
    Generate three random outfits (Top + Bottom or Dress + Jacket).
    """
    outfits = []
    for _ in range(3):
        top_or_dress = df[df['Type'].isin(['Top', 'Dress'])].sample(1).to_dict('records')[0]
        if top_or_dress['Type'] == 'Dress':
            bottom_or_jacket = {
                'ID': None,
                'Type': 'Jacket',
                'Color': 'N/A',
                'Size': 'N/A',
                'Image': 'N/A'
            }
            outfits.append((bottom_or_jacket, top_or_dress))
        else:
            bottom_or_jacket = df[df['Type'] == 'Bottom'].sample(1).to_dict('records')[0]
            outfits.append((top_or_dress, bottom_or_jacket))
    return outfits

# Mostrar curva de aprendizaje
def plot_learning_curve(feedback_db):
    """
    Plot the learning curve based on the feedback database.
    """
    feedback_df = pd.DataFrame(feedback_db)
    feedback_df['Iteration'] = range(1, len(feedback_db) + 1)
    feedback_df['Cumulative Feedback'] = feedback_df['Feedback'].cumsum()
    feedback_df['Cumulative Proportion'] = feedback_df['Cumulative Feedback'] / feedback_df['Iteration']

    plt.figure(figsize=(8, 6))
    plt.plot(feedback_df['Iteration'], feedback_df['Cumulative Proportion'], marker='o')
    plt.title("Learning Curve: Proportion of Positive Feedback")
    plt.xlabel("Iteration")
    plt.ylabel("Cumulative Positive Feedback Proportion")
    plt.grid()
    plt.show()

# Mostrar outfits y calificar
def display_outfits_interactive(outfits):
    """
    Display three outfits with buttons for rating (✔ and ✘) and generate/terminate buttons.
    """
    output_boxes = []
    session_feedback = []

    def rate_outfit(outfit_index, feedback):
        """
        Handle the rating for a specific outfit.
        """
        top_or_dress, bottom_or_jacket = outfits[outfit_index]
        feedback_db.append({
            'Top_or_Dress_ID': top_or_dress['ID'],
            'Bottom_or_Jacket_ID': bottom_or_jacket.get('ID'),
            'Feedback': feedback
        })
        print(f"Rated outfit {outfit_index + 1}: {'✔' if feedback else '✘'}")
        regenerate_outfits(None)

    for index, (top, bottom) in enumerate(outfits):
        out = Output()

        with out:
            # Display outfit
            fig, axes = plt.subplots(2, 1, figsize=(3, 6))  # Adjusted to make photos smaller

            # Top or Jacket
            if os.path.exists(top['Image']):
                img = mpimg.imread(top['Image'])
                axes[0].imshow(img)
                axes[0].set_title(f"{top['Type']} ({top['Color']}, {top['Size']})")
            else:
                axes[0].text(0.5, 0.5, "Image not found", ha='center', va='center', fontsize=10, color='red')
                axes[0].set_title(f"{top['Type']} ({top['Color']}, {top['Size']})")
            axes[0].axis('off')

            # Bottom or Dress
            if os.path.exists(bottom['Image']):
                img = mpimg.imread(bottom['Image'])
                axes[1].imshow(img)
                axes[1].set_title(f"{bottom['Type']} ({bottom['Color']}, {bottom['Size']})")
            else:
                axes[1].text(0.5, 0.5, "Jacket not available" if bottom['Type'] == "Jacket" else "Image not found",
                             ha='center', va='center', fontsize=10, color='red')
                axes[1].set_title(f"{bottom['Type']} ({bottom['Color']}, {bottom['Size']})")
            axes[1].axis('off')

            plt.show()

        # Buttons for rating
        yes_button = Button(description="✔", layout=Layout(width='50px'))
        no_button = Button(description="✘", layout=Layout(width='50px'))

        yes_button.on_click(lambda _: rate_outfit(index, True))
        no_button.on_click(lambda _: rate_outfit(index, False))

        # Add to layout
        output_boxes.append(VBox([out, HBox([yes_button, no_button])]))

    # Generate New and Terminate Buttons
    generate_button = Button(description="Generate New Outfit", layout=Layout(width='200px'))
    terminate_button = Button(description="Terminate", layout=Layout(width='200px'))

    def regenerate_outfits(_):
        new_outfits = generate_outfits(df)
        display_outfits_interactive(new_outfits)

    def terminate_session(_):
        plot_learning_curve(feedback_db)

    generate_button.on_click(regenerate_outfits)
    terminate_button.on_click(terminate_session)

    display(VBox([
        HBox([generate_button, terminate_button]),
        HBox(output_boxes)
    ]))

# Mostrar los outfits iniciales
initial_outfits = generate_outfits(df)
display_outfits_interactive(initial_outfits)
```
