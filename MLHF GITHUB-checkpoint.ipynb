# RENINFORCEMENT 

```python
import pandas as pd
import os
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Crear el DataFrame inicial con distribución específica
def create_dataframe():
    """
    Create a DataFrame with specific counts for tops, bottoms, dresses, and jackets.
    """
    image_path = r'C:\Users\POspina\OneDrive - SLB\Desktop\Backup\Tesis\Fotos'

    # Generate data for each type
    data = []

    # Tops (51 items)
    for i in range(1, 52):
        data.append({
            'ID': i,
            'Type': 'Top',
            'Color': random.choice(['White', 'Black', 'Red', 'Blue']),
            'Size': random.choice(['S', 'M', 'L', 'XL']),
            'Image': os.path.join(image_path, f"Top ({i}).png")
        })

    # Bottoms (22 items)
    for i in range(1, 23):
        data.append({
            'ID': 51 + i,
            'Type': 'Bottom',
            'Color': random.choice(['White', 'Black', 'Red', 'Blue']),
            'Size': random.choice(['S', 'M', 'L', 'XL']),
            'Image': os.path.join(image_path, f"Bottom ({i}).png")
        })

    # Dresses (15 items)
    for i in range(1, 16):
        data.append({
            'ID': 73 + i,
            'Type': 'Dress',
            'Color': random.choice(['White', 'Black', 'Red', 'Blue']),
            'Size': random.choice(['S', 'M', 'L', 'XL']),
            'Image': os.path.join(image_path, f"Dress ({i}).png")
        })

    # Jackets (29 items)
    for i in range(1, 30):
        data.append({
            'ID': 88 + i,
            'Type': 'Jacket',
            'Color': random.choice(['White', 'Black', 'Red', 'Blue']),
            'Size': random.choice(['S', 'M', 'L', 'XL']),
            'Image': os.path.join(image_path, f"Jacket ({i}).png")
        })

    return pd.DataFrame(data)

df = create_dataframe()
feedback_db = []

# Función para generar outfits
def generate_outfits(df):
    """
    Generate three random outfits (Top + Bottom + Jacket or Dress + Jacket).
    """
    outfits = []
    for _ in range(3):
        top_or_dress = df[df['Type'].isin(['Top', 'Dress'])].sample(1).to_dict('records')[0]
        if top_or_dress['Type'] == 'Dress':
            jacket = df[df['Type'] == 'Jacket'].sample(1).to_dict('records')[0]
            outfits.append((jacket, top_or_dress))
        else:
            bottom = df[df['Type'] == 'Bottom'].sample(1).to_dict('records')[0]
            outfits.append((top_or_dress, bottom))
    return outfits

# Mostrar outfits en columnas
def display_outfits(outfits):
    """
    Display outfits in a grid with 3 columns using matplotlib.
    """
    fig, axes = plt.subplots(2, 3, figsize=(12, 8))  # 2 filas, 3 columnas

    for col, outfit in enumerate(outfits):
        # Mostrar primer artículo (top o chaqueta)
        item1 = outfit[0]
        if os.path.exists(item1['Image']):
            img = mpimg.imread(item1['Image'])
            axes[0, col].imshow(img)
            axes[0, col].set_title(f"{item1['Type']} ({item1['Color']}, {item1['Size']})")
        else:
            axes[0, col].text(0.5, 0.5, "Image not found", ha='center', va='center', fontsize=10, color='red')
            axes[0, col].set_title(f"{item1['Type']} ({item1['Color']}, {item1['Size']})")
        axes[0, col].axis('off')

        # Mostrar segundo artículo (bottom o dress)
        item2 = outfit[1]
        if os.path.exists(item2['Image']):
            img = mpimg.imread(item2['Image'])
            axes[1, col].imshow(img)
            axes[1, col].set_title(f"{item2['Type']} ({item2['Color']}, {item2['Size']})")
        else:
            axes[1, col].text(0.5, 0.5, "Image not found", ha='center', va='center', fontsize=10, color='red')
            axes[1, col].set_title(f"{item2['Type']} ({item2['Color']}, {item2['Size']})")
        axes[1, col].axis('off')

    plt.tight_layout()
    plt.show()

# Calificar outfits
def rate_outfits(outfits):
    """
    Prompt the user to rate each outfit and store feedback.
    """
    for idx, outfit in enumerate(outfits):
        rating = int(input(f"Rate outfit {idx + 1} (1 for Like, 0 for Dislike): "))
        feedback_db.append({
            'Outfit_ID': idx + 1,
            'Feedback': rating
        })

# Mostrar curva de aprendizaje
def plot_learning_curve(feedback_db):
    """
    Plot the learning curve based on the feedback database.
    """
    feedback_df = pd.DataFrame(feedback_db)
    feedback_df['Iteration'] = range(1, len(feedback_db) + 1)
    feedback_df['Cumulative Feedback'] = feedback_df['Feedback'].cumsum()
    feedback_df['Cumulative Proportion'] = feedback_df['Cumulative Feedback'] / feedback_df['Iteration']

    plt.figure(figsize=(8, 6))
    plt.plot(feedback_df['Iteration'], feedback_df['Cumulative Proportion'], marker='o')
    plt.title("Learning Curve: Proportion of Positive Feedback")
    plt.xlabel("Iteration")
    plt.ylabel("Cumulative Positive Feedback Proportion")
    plt.grid()
    plt.show()

# Ciclo de evaluación
def evaluate_outfits():
    """
    Main loop to evaluate outfits and optionally terminate.
    """
    while True:
        outfits = generate_outfits(df)
        display_outfits(outfits)
        rate_outfits(outfits)
        continue_rating = input("Do you want to rate more outfits? (yes/no): ").strip().lower()
        if continue_rating != 'yes':
            plot_learning_curve(feedback_db)
            break

# Ejecutar evaluación
evaluate_outfits()
```
